### Important Code

```python
# If the name includes Mrs Change title to Mrs, Change gender to Female & change marital_status to Married
# If the name includes Mr Change title to Mr & Change gender to Male 
# If the name includes Miss Change title to Miss, Change gender to Female & change marital_status to Single
# If the name includes Miss Change title to Ms, Change gender to Female & change marital_status to Single


employee.loc[employee['Name'].str.contains('Mrs '), ['Title', 'Gender','Marital_Status']] = ['Mrs', 'Female','Married']
employee.loc[employee['Name'].str.contains('Mr '), ['Title', 'Gender']] = ['Mr', 'Male']
employee.loc[employee['Name'].str.contains('Ms '), ['Title', 'Gender','Marital_Status']] = ['Miss', 'Female','Single']
employee.loc[employee['Name'].str.contains('Miss '), ['Title', 'Gender','Marital_Status']] = ['Miss', 'Female','Single']
```
