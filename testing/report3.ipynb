```python
# Do necessary imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import holidays, salary_dictionary, train-attendance, train-leaves, train-salary csv files in data folder from pd
holidays = pd.read_csv('../data/holidays.csv')
salary_dictionary = pd.read_csv('../data/salary_dictionary.csv')
train_attendance = pd.read_csv('../data/train-attendance.csv')
train_leaves = pd.read_csv('../data/train-leaves.csv')
train_salary = pd.read_csv('../data/train-salary.csv')

cleaned_emp = pd.read_csv('../employee_preprocess_200304N.csv')
report_salary = pd.read_csv('../testing/report_salary_df.csv')
```

```python
# create a new df showing net salary per month per employee
per_month = train_salary.groupby(['Employee_No', 'month', 'year'])['Net Salary'].sum().reset_index()
per_month
```

```python
# calculate the average net salary per month per employee
avg_salary = per_month.groupby(['Employee_No'])['Net Salary'].mean().reset_index()
avg_salary
```

```python
# merge train_salary and cleaned_emp on Employee_NO. keep only Net Salary, Total_Earnings_0, Total_Earnings_2, Employment_Type, Religion, Designation
report_salary = pd.merge(avg_salary, cleaned_emp[['Employee_No','Employment_Type', 'Religion', 'Designation','Employment_Category','Year_of_Birth']], on='Employee_No', how='left') 
```

```python
report_salary = report_salary[['Employee_No', 'Net Salary','Employment_Type', 'Religion', 'Designation']]
report_salary
```

```python
report_salary.isna().sum()
```

```python
report_salary.dropna(inplace=True)
```
