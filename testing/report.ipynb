```python
df = train_attendance.copy()
```

```python
# Convert Shift_Start and Shift_End columns to datetime format
df['Shift_Start'] = pd.to_datetime(df['Shift_Start'])
df['Shift_End'] = pd.to_datetime(df['Shift_End'])

# Calculate the shift_time by subtracting Shift_Start from Shift_End
df['shift_time'] = df['Shift_End'] - df['Shift_Start']

# Extract the duration in hours from the shift_time
df['shift_time'] = df['shift_time'].dt.total_seconds() / 3600
```

```python
df
```

```python
def convert_datetime(df):
    try:
        df['in_time'] = pd.to_datetime(df['in_time'])
        df['out_time'] = pd.to_datetime(df['out_time'])
    except ValueError as e:
        print(f"Error converting {e}")
    return df

df = convert_datetime(df.copy())
```

```python
df = df.loc[df['out_time'] != '24:50:00']
```

```python
df['out_time'] = pd.to_datetime(df['out_time'])
```

```python
df
```

```python
def recalculate_hourly_time(df):
    try:
        df['Hourly_Time'] = (df['out_time'] - df['in_time']).dt.total_seconds() / 3600
    except TypeError as e:
        print(f"Error converting {e}")
    return df

df = recalculate_hourly_time(df.copy())
```

```python
df
```

```python
def calculate_overtime(df):
    try:
        df['over_time'] = df['Hourly_Time'] - df['shift_time']
    except TypeError as e:
        print(f"Error converting {e}")
    return df

df = calculate_overtime(df.copy())
```

```python
df
```

```python
def create_project_df(df):
    project_df = df.groupby('project_code')['shift_time'].sum()
    project_df = pd.DataFrame(project_df).reset_index()
    return project_df

project_df = create_project_df(df.copy())
```

```python
project_df
```

```python
def create_emp_df(df):
    emp_df = df.groupby('Employee_No')['shift_time'].sum()
    emp_df = pd.DataFrame(emp_df).reset_index()
    return emp_df

emp_df = create_emp_df(df.copy())
```

```python
emp_df
```

```python
def create_ovr_df(df):
    ovr_df = df.groupby('Employee_No')['over_time'].sum()
    ovr_df = pd.DataFrame(ovr_df).reset_index()
    return ovr_df

ovr_df = create_ovr_df(df.copy())


ovr_df
```
