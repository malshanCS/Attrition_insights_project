```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
attendance = pd.read_csv("data/train-attendance.csv")
attendance.head()
```

```python

# Get all rows affiliated with a certain Employee_No
attendance.loc[attendance['Employee_No'] == 1272]
```

```python
attendance.loc[attendance['out_date'] == '0000-00-00', 'out_date'] = '3/1/2022'
```

```python
attendance['out_date'] = pd.to_datetime(attendance['out_date'])
```

```python
# Create a new DataFrame with the latest out_date for each Employee_No
latest_dates = attendance.loc[attendance.groupby('Employee_No')['out_date'].idxmax()]
```

```python
latest_dates.loc[latest_dates['Employee_No']==1272]
```

```python
employee_cleaned = pd.read_csv('employee_cleaned.csv')
employee_cleaned
```

```python

# Assuming 'employee_cleaned' and 'attendance' are the DataFrames containing the data
new_df = pd.merge(employee_cleaned[['Employee_No', 'Date_Resigned']], latest_dates[['Employee_No', 'out_date']], on='Employee_No', how='left')
```

```python
new_df
```

```python
temp = new_df.loc[new_df['Date_Resigned'] != '0000-00-00']
```

```python
temp 
```

```python
# convert Date_Resigned to datetime
temp['Date_Resigned'] = pd.to_datetime(temp['Date_Resigned'])
```

```python
# if Date_Resigned is after out_date, then set Date_Resigned to out_date
temp.loc[temp['Date_Resigned'] > temp['out_date'], 'Date_Resigned'] = temp['out_date']
```

```python
final = pd.merge(employee_cleaned[['Employee_No', 'Date_Resigned']], temp[['Employee_No', 'Date_Resigned']], on='Employee_No', how='left')
```
